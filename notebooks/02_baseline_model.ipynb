# 02 ベースラインモデル

**目的**: シンプルなモデルで「最初のお手本精度（ベースライン）」を作る。

ベースライン = あとから工夫したモデルと比べるためのスタート地点。

---

## このノートブックでやること

1. 学習データを読み込む
2. 前処理（欠損値・カテゴリ変数の変換）
3. モデルを学習して精度を測る
4. モデルと前処理の情報を保存する

---

## ヒント

### データ読み込み
- `from analysis_titanic.load_data import load_train`
- `from analysis_titanic.paths import MODELS_DIR`

### 前処理の考え方
機械学習は **数字の表** を入力する必要がある。

| やること | 例 |
|---|---|
| 欠損値を埋める | `age` → 中央値、`embarked` → 最頻値 |
| カテゴリを数値化 | `sex` → 0/1、`embarked` → one-hot |
| 不要列を除く | `id` は特徴量に使わない |

※ 学習時に決めた値（中央値など）は、後でテストデータにも **同じ値** を使う。

### モデルの例
- `LogisticRegression` … シンプルで解釈しやすい
- `RandomForestClassifier` … 少し複雑だが当たりやすいことも

### 精度の測り方
- **交差検証（Cross Validation）** で評価すると安心
- 指標: `accuracy`（正解率）
- `sklearn.model_selection.cross_val_score` が使える

### 再現性
- `RANDOM_SEED = 42` を決めて `random_state` に渡す

### 保存
- `joblib.dump(...)` で `models/baseline_model.joblib` に保存
- モデルだけでなく **前処理の情報**（中央値・列名など）も一緒に保存すると便利

### 参考: 保存する中身の例
```python
{
    "model": 学習済みモデル,
    "age_median": ...,
    "embarked_mode": ...,
    "feature_columns": [...],
}
```
